#Clase 3: Data Munging / Data Wrangling

```python
import pandas as pd
import numpy as np
```

```python
x = {"Company" : ['Ford', 'Ford', 'VW', 'BMW', 'Cooper', 'Cooper'],
     "Stars": [1, 2, np.nan, 3, 2, 2],
     "Weight": [2, 4, 2, 2, 3, None],
     'Origin': ["Mexico", "China", "China", "Mexico", "UK", None],
     "Length": [40, 30, np.nan, 56, pd.NaT, 40]
     }
```

```python
type(x)
```

```python
df = pd.DataFrame(data = x)
df
```

```python
df.isnull().any()
```

```python
ndf = df.copy()
ndf.shape
```

```python
ndf.dropna()
```

```python
ndf.isnull().any(axis=1)
```

```python
ndf
```

```python
ndf.dropna()
```

```python
ndf.dropna(inplace = True)
```

```python
ndf
```

```python
sm = ndf.Stars.median()
sm
```

```python
ndf['Stars'] = ndf['Stars'].fillna(value = sm)
ndf
```

```python
wm = ndf.Weight.mean()
om = ndf.Origin.mode()[1]
lm = ndf.Length.median()
```

```python
ndf.Weight = ndf.Weight.fillna(wm)
ndf.Origin = ndf.Origin.fillna(om)
ndf.Length = ndf.Length.fillna(lm)
```

```python
ndf
```

```python
pd.pivot_table(ndf,
               index = "Company",
               values = 'Weight',
               aggfunc = ['mean', 'median'])
```
